```python
import os
import numpy as np
import pandas as pd

# =====================================================
# DATASET
# =====================================================

run_path = "."

# =====================================================
# LOAD LABELS
# =====================================================

y_train = np.load(
    os.path.join(run_path, "y_train_fold0.npy"),
    allow_pickle=True
)

y_test = np.load(
    os.path.join(run_path, "y_test_fold0.npy"),
    allow_pickle=True
)

# =====================================================
# CONCAT LABELS
# =====================================================

y_all = np.concatenate(
    [y_train, y_test],
    axis=0
)

# =====================================================
# DISPLAY
# =====================================================

print("Unique labels:")
print(np.unique(y_all))

print("\nDistribution:")
print(pd.Series(y_all).value_counts())

# =====================================================
# SAVE CSV
# =====================================================

labels_df = pd.DataFrame({
    "label": y_all
})

csv_path = os.path.join(
    run_path,
    "labels_info.csv"
)

labels_df.to_csv(
    csv_path,
    index=False
)

print(f"\nSaved: {csv_path}")
```
